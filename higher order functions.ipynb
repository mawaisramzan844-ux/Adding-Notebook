# Higher order functions

```python
celsius = [0,37,23,45]
fahrenhiet = list(map(lambda c: (c * 9/5)+32, celsius))

print(fahrenhiet)
```

```python
from functools import reduce

def multiply(x,y):
    return x* y

numbers = [1,2,3,4,5,6]
products =reduce(multiply,numbers)

print(products)
```

```python
def my_decorator(func):

    def wrapper(*arg, **kwarg):
        print("Hello!")
        result = func(*arg,**kwarg)
        print("Good bye!")
        return result
    
    return wrapper

@my_decorator
def say_hello():
    pass

say_hello()

```
