```python
def calculate_cost (price, quality):
    return price * quality

price = float(input("Enter price :"))
quality = int(input("Enter quality :"))
total = calculate_cost(price , quality)

print("Your total cost is :",total)
```

```python
def temp_converter (temp_value, unit):
    if unit == "C":
        return (temp_value - 32) * 5/9  # Celsius to Fahrenheit 
    elif unit == "F":
        return (temp_value * 9/5) + 32    # Fahrenheit to celsius
    else:
        return ("Invalid choice! Please select ('C'/'F')")
    
temp_value = float(input("Enter Temprature : "))
unit = input("Enter choice (C/F) :").upper()

result = temp_converter(temp_value , unit)

print("Your is temprature is :",result)
```

```python
def apply_disscount(original_price, disscount_percentage):
    return (original_price -(original_price * disscount_percentage / 100)) # formula to find final price

original_price = float(input("Enter original price :"))
disscount_percentage = float(input("Enter disscount_percentage :"))
final_price = apply_disscount(original_price, disscount_percentage)

print("Original price is :",original_price)
print("Disscount percentage is :",disscount_percentage)
print("Final price is :",final_price)
```
