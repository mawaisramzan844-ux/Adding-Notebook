# Convert tuple into list

```python
nums = (1,2,3,4,5,6)
# to convert tuple into list
list_nums = list(nums)
# now append
list_nums.append(4)
print(list_nums)
```

```python
element = (11,22,33,44,55)

list_element = list(element)
# list_element[2:4]

print(list_element[0:-1])
```

# Dictionary

```python
list = {
    "Name" : "Awais",
    "Age" : 19,
    "City" : "Bahawalpur",
}
print(list["Name"])
```

```python
items = {
    "Glass",
    "Ball",
    "bat"
}
print(len(items))
```

```python
dict = {
    "Name": "Awais",
    "City": "Bahawalpur"
}
dict["age"] = 18

print(dict)
```

```python
val = {
    "a":12,
    "b":14,
    "c":10
    }
val_sum = sum((val.values()))
print((val_sum))
```

```python
marks = {
    "maths": int(input("Enter your marks in maths: ")),
    "physics": int(input("Enter your marks in physics: ")),
    "computer": int(input("Enter your marks in computer: ")),
}
total_marks = sum(marks.values())
avg = total_marks/(len(marks))

print("Average is :",avg)
```
